```python
import numpy as np
import pandas as pd
import math
import numbers
import matplotlib.pyplot as plt
from pandas.tools.plotting import scatter_matrix
import random
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.stats import chisquare
%matplotlib inline
```

```python
bankChurn = pd.read_csv('./bankChurn.csv', encoding='utf-8')
externalData = pd.read_csv('./ExternalData.csv', encoding='utf-8')
df = pd.merge(bankChurn, externalData, on='CUST_ID')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
columns = set(list(df.columns))
# 去除目标变量
columns.remove('CHURN_CUST_IND')
```

```python
# 将数值型属性和类别型属性分别存放
numericCols = []
stringCols = []
for var in columns:
    x = list(set(df[var]))
    x = [i for i in x if i==i] # 去除噪声，即nan类型的数据
    if isinstance(x[0], numbers.Real):
        numericCols.append(var)
    elif isinstance(x[0], str):
        stringCols.append(var)
    else:
        print('The type of ' + var + ' cannot be determined.')
```

```python
print('length of numericCols: ', len(numericCols))
print('length of stringCols: ', len(stringCols))
```

```python
# 数值型属性可视化分析
def NumVarPerf(df, col, target, truncation=False):
    '''
    :param df: 数据集，包含自变量和因变量
    :param col: 自变量（属性变量）
    :param target: 因变量，0或1
    :param filepath: 存储柱状图的文件路径
    :param truncation: 是否需要对某些离群点截断
    :return: 返回描述性统计值
    '''
    # 抽取目标变量和特定的属性变量, df[col]== df[col]会去除nan的样本
    validDf = df.loc[df[col]== df[col]][[col, target]]
    # 有效样本占总数据集中样本的比例
    validRcd = validDf.shape[0] * 1.0 / df.shape[0]
    validRcdFmt = "%.2f%%" % (validRcd * 100)
    # 每一列数据的描述性统计值
    descStats = validDf[col].describe()
    mu = "%.2e" % descStats['mean']
    std = "%.2e" % descStats['std']
    maxVal = "%.2e" % descStats['max']
    minVal = "%.2e" % descStats['min']
    # 展示流失与为流失的分布情况
    x = validDf.loc[validDf[target] == 1][col]
    y = validDf.loc[validDf[target] == 0][col]
    xweights = 100.0 * np.ones_like(x) / x.size
    yweights = 100.0 * np.ones_like(y) / y.size
    # 在需要trancation时，对95分位点进行截断
    if truncation == True:
        pcnt95 = np.percentile(validDf[col], 95)
        x = x.map(lambda x: min(x, pcnt95))
        y = y.map(lambda y: min(y, pcnt95))
    fig, ax = plt.subplots()
    ax.hist(x, weights=xweights, alpha=0.5, label='Attrition')
    ax.hist(y, weights=yweights, alpha=0.5, label='Retained')
    titleText = 'Histgram of ' + col + '\n' + 'valid pcnt = ' + validRcdFmt + ', Mean = ' + mu + ', Std = ' + std + '\n max = ' + maxVal + ', min = ' + minVal
    ax.set(title = titleText, ylabel='% of Dataset in Bin')
    ax.margins(0.05)
    ax.set_ylim(bottom=0)
    ax.legend(loc='upper right')
    #figSavePath = 'Num_' + str(col) + '.png'
    #plt.savefig(figSavePath)
    plt.show()
```

```python
for var in numericCols:
    NumVarPerf(df, var, 'CHURN_CUST_IND', True)
```

```python
# 方差检验 anova test
anova_result = anova_lm(ols('ASSET_MON_AVG_BAL~CHURN_CUST_IND', df).fit())
anova_result
```

```python
# 类别型属性可视化分析
def CharVarPerf(df, col, target):
    '''
    :param df: 数据集，包含自变量和因变量
    :param col: 自变量（属性变量）
    :param target: 因变量，0或1
    '''
    validDf = df.loc[df[col] == df[col]][[col, target]]
    validRcd = validDf.shape[0] * 1.0 / df.shape[0]
    redNum = validDf.shape[0]
    validRcdFmt = "%.2f%%" % (validRcd * 100)
    freqDict = {}
    churnRateDict = {}
    for v in set(validDf[col]):
        vDf = validDf.loc[validDf[col] == v]
        freqDict[v] = vDf.shape[0] * 1.0 / redNum
        churnRateDict[v] = sum(vDf[target]) * 1.0 / vDf.shape[0]
    descStats = pd.DataFrame({'percent': freqDict, 'churn rate': churnRateDict})
    fig = plt.figure()
    ax = fig.add_subplot(111)
    # 创建另一个ax对象，并于第一个ax对象共享x轴
    ax2 = ax.twinx()
    plt.title('The percentage and churn rate for ' + col + '\n valid pcnt = ' + validRcdFmt)
    descStats['churn rate'].plot(kind='line', color='red', ax=ax)
    descStats.percent.plot(kind='bar', color='blue', ax=ax2, width=0.2, position=1)
    ax.set_ylabel('churn rate')
    ax2.set_ylabel('percentage')
    #figSavePath = 'Str_' + str(col) + '.png'
    #plt.savefig(figSavePath)
    plt.show()
```

```python
for var in stringCols:
    CharVarPerf(df, var, 'CHURN_CUST_IND')
```

```python
# 卡方检验
chisqDf = df[['GENDER_CD','CHURN_CUST_IND']]
grouped = chisqDf['CHURN_CUST_IND'].groupby(chisqDf['GENDER_CD'])
count = list(grouped.count())
churn = list(grouped.sum())
chisqTable = pd.DataFrame({'total': count, 'churn': churn})
chisqTable['expected'] = chisqTable['total'].map(lambda x: round(x*0.101))
chisqValList = chisqTable[['churn', 'expected']].apply(lambda x: (x[0]-x[1])**2 / x[1], axis=1)
chisqVal = sum(chisqValList)
print('chisqVal: ', chisqVal)
# 从上面计算得到，低于0.05且自由度为2的卡方值是5.99，而计算得到的卡方值是32.66，因此性别属性是显著的。
# 我们也可以直接使用chisquare函数来计算
chisquare(chisqTable['churn'], chisqTable['expected'])
```
